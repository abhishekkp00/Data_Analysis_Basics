```python
import matplotlib.pyplot as plt
```

```python
x = [1, 2, 3, 4, 5]
y = [1, 4, 9, 16, 25]
plt.plot(x, y)
plt.xlabel('X Axis')
plt.ylabel('Y axis')
plt.title('Basic Line Plot')
plt.show()
```

```python
x = [1, 2, 3, 4, 5]
y = [1, 4, 9, 16, 25]

#Create customised line plot
plt.plot(x, y, color = 'red', linestyle = '--')
```

```python
#Multiple pLots
#Sample data
x = [1, 2, 3, 4, 5]
y = [1, 4, 9, 16, 25]
z = [2, 4, 6, 8, 10]

plt.figure(figsize = (9,5))
plt.subplot(2,2,1)
plt.plot(x, y, color = 'green')
plt.title('plot 1')

plt.figure(figsize = (9,5))
plt.subplot(2, 2, 2)
plt.plot(y, z, color = 'red')
plt.title('plot 2')

plt.figure(figsize = (9,5))
plt.subplot(2,2, 3)
plt.plot(x, z, color = 'blue')
plt.title('plot 3')

plt.figure(figsize = (9,5))
plt.subplot(2,2, 4)
plt.plot(y,x , color = 'yellow')
plt.title('plot 4')
```

```python
#bar plot
category = ['A', 'B', 'C', 'D', 'E']
values= [5, 6, 7, 8, 9]

#creating the bar plot
plt.bar(category, values, color = 'purple')
plt.xlabel('X Axis')
plt.ylabel('Y axis')
plt.title('Basic Line Plot')
plt.show()
```

```python
#Histogram
data = [1, 2, 2, 3, 3, 3, 4, 4, 4, 4]
plt.hist(data, bins = 4, color = 'orange', edgecolor = 'black')
```

```python
#SCATTER PLOT
x = [1, 2, 3, 4]
y = [2, 3, 4, 5]
plt.scatter(x, y, color = 'blue', marker = 'x')
```

```python
# PIE CHARTS
labels = ['A', 'B', 'C', 'D']
sizes = [20, 30, 40, 50]
colors = ['gold', 'yellow', 'blue', 'pink']
explode = (0.2, 0, 0, 0)

plt.pie(sizes, explode = explode, labels = labels, colors = colors, autopct = "%1.1f%%", shadow = 'True')
```
