```python
import sqlite3
```

```python
connection = sqlite3.connect('example.db')
connection
```

```python
cursor = connection.cursor()
```

```python
cursor.execute(''' 
    Create table If Not Exists Employees(
        Id Integer Primary Key,
        Name varchar Not Null,
        Age integer,
        Department text
    )
''')
```

```python
connection.commit
```

```python
cursor.execute(''' Insert Into Employees(Id, Name, Age, Department) Values
    (101, "Abhishek", 32, "Data Scientist")
''')
```

```python
cursor.execute(''' Insert Into Employees(Id, Name, Age, Department) Values
    (102, "Ankita", 24, "Milu")
''')
```

```python
connection.commit()
```

```python
cursor.execute('Select * from Employees')
rows = cursor.fetchall()
```

```python
for row in rows:
    print(row)
```

```python
##Update the data in table

cursor.execute(''' Update Employees set age = 34 where name = 'Abhishek'
''')
connection.commit()
```

```python
cursor.execute('Select * from Employees')
rows = cursor.fetchall()
```

```python
for row in rows:
    print(row)
```

```python
##Daleting a data from the table
cursor.execute('''
Delete from employees where name = "Ankita"
''')
connection.commit()
```
