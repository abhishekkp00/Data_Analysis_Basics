```python
df = pd.read_csv("sales_data_sample.csv", encoding="latin1")
df.head(5)
```

```python
df.describe()
```

```python
df.dtypes
```

```python
#Handling Missing Values
df.isnull().any()
```

```python
df.isnull().sum() #sum of missing values
```

```python
#fill missing with zero
df.fillna(0)
```

```python
df.isnull().sum() #sum of missing values
```

```python
#filling missing number with nonof the column
df['Address_fillNA'] = df['ADDRESSLINE2'].fillna("Non")
```

```python
df.isnull().sum() #sum of missing values
```

```python
#Renaming column name
df = df.rename(columns={"STATUS": 'NEW STATUS'})
```

```python
df.head(5)
```

```python
#change datatype
df['SALES_NEW'] = df["SALES"].astype(int)
```

```python
df.head(5)
```

```python
#Apply function
df['PRICE_NEW'] = df['PRICEEACH'].apply(lambda x: x*2)
```

```python
df.head(5)
```

```python
#Data AGgregating and GRouping
grouped_mean = df.groupby("COUNTRY")['SALES'].mean()
```

```python
print(grouped_mean)
```

```python
#AGGREGATE MULTIPLE FUNCTION
grouped_agg = df.groupby('COUNTRY')['SALES'].agg(['mean','sum','count'])
```

```python
print(grouped_agg)
```

```python
#Merging and joining dataframes
#Create sample dataframe
df1 = pd.DataFrame({'Keys': [1, 2, 3, 4], 'Values': ['a', 'b', 'c', 'd']})
df2 = pd.DataFrame({'Keys': [1, 2, 3, 8], 'Values': ['b', 'b', 'c', 'e']})
```

```python
df1
```

```python
df2
```

```python
#Merge dataframe on key columns
pd.merge(df1, df2, on="Keys", how="inner")
```

```python
pd.merge(df1, df2, on="Keys", how="outer")
```

```python
pd.merge(df1, df2, on="Keys", how="left")
```

```python
pd.merge(df1, df2, on="Keys", how="right")
```
